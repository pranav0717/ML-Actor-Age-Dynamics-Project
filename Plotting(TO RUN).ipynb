STEP 2:

PULLING DATA FROM CREATED DATA FILES AND SHOWING SAMPLES

```python
#creating actresses dataframe
actresses_df = pd.read_csv('/content/drive/My Drive/actresses_df.cvs', sep=',')
actresses_df
```

```python
#creating actors dataframe
actor_df = pd.read_csv('/content/drive/My Drive/actors.cvs' , sep=',')
actor_df
```

STEP 3:

INPUT PARSING

```python
#input parsing
x = input("name of actor? (Case sensetive)")
find = actor_df[actor_df["primaryName"] == x]
if find.empty:
    print("Actor not in dataset.")
else:
  tot = find['tconst']
  actresses_df_new = actresses_df[actresses_df['tconst'].isin(tot)].copy()
```

STEP 4:

AGE CALCULATION AND NaN EDGE CASE HANDLING

```python
#dataframe mainpulation
unique_actresses_df = actresses_df_new.drop_duplicates(subset=["nconst", "primaryName", "birthYear", "primaryProfession"])
unique_actresses_df = unique_actresses_df.replace("\\N", pd.NA).fillna(method='ffill')

#age calculation
actor_age = pd.to_numeric(find["birthYear"].replace("\\N", pd.NA)).iloc[0]

#nan values edge case handling
if np.isnan(actor_age):
    print("actor age not found, please try again")
    actor_age = 0
```

```python
actor_age
```

```python
#plotting age of actresses with respect to age of actors
fig = go.Figure(go.Scatter(x=unique_actresses_df["primaryName"], y=  actor_age - pd.to_numeric(unique_actresses_df["birthYear"], errors='coerce')))
fig.show()
```
